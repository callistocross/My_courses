**Задание 1.**
Напишите программу, которая рисует пунктирную линию.

```python
import turtle as t
t.shape('circle')

for _ in range(10):
    t.stamp()
    t.penup()
    t.forward(30)
    t.pendown()

t.mainloop()
```

**Задание 2.** Напишите программу, которая рисует прямоугольник с точкой в каждом углу.

```python
import turtle as t
t.shape('circle')
t.pensize(3)
a, b = 100, 150
for i in [a, b] * 2:
  t.right(90)
  t.forward(i)
  t.stamp()

t.mainloop()
```

**Задание 3.** Напишите программу для рисования паутины в соответствии с примером. Программа должна считывать количество лучей паутины, число n.

```python
import turtle as t
n = 20
angle = 360 // n
side = 100
t.pensize(3)
t.shape('circle')
t.stamp()

t.shape('arrow')
for _ in range(n):
    t.forward(side)
    t.stamp()
    t.backward(side)
    t.left(angle)


t.mainloop()
```

**Задание 5.** Напишите программу, которая рисует циферблат часов в соответствии с образцом.

```python
import turtle as t
n = 12
angle = 360 // n
side = 150
t.Screen().bgcolor('CadetBlue3')
t.pensize(3)
t.shape('turtle')
t.stamp()

for _ in range(n):
    t.penup()
    t.forward(7*(side//10))
    t.pendown()
    t.forward(2*(side//10))
    t.penup()
    t.forward(1*(side//10))
    t.stamp()
    t.backward(10*(side//10))
    t.right(angle)
    
t.mainloop()
```

**Задание 7.** Напишите программу, которая рисует узор в соответствии с образцом.

```python
import turtle as t
d = 3
Colors = ['Blue', 'Yellow', 'Green', 'Magenta', 'DarkOrange', 'Red']
r = 0
for i in range(10, 150, 3):
    t.pensize(d)
    t.pencolor(Colors[r])
    t.left(45)
    t.forward(i)
    r += 1
    if r >= len(Colors): r = 0
    d += 0.2   
t.mainloop()
```

**Задание 8.** Напишите программу, которая рисует звезду, показанную на рисунке. Такую звезду можно создать из двух треугольников. Однако их невозможно нарисовать непрерывной линией, поэтому перо нужно будет поднять для перехода ко второму треугольнику.

```python
import turtle as t
t.pensize(3)
side = 80
t.goto(side,0)
t.goto((1/2)*side,side)
t.goto(0,0)
t.penup()
t.goto(0, (2/3)*side)
t.pendown()
t.goto(side,(2/3)*side)
t.goto((1/2)*side,-(1/3)*side)
t.goto(0,(2/3)*side)
t.mainloop()
```

**Задание 9.** Напишите программу, которая рисует изображение в соответствии с образцом.

```python
import turtle as t

t.pensize(2)
n = 12
a = 300 / n
t.penup()
t.hideturtle()
if n % 2 == 0:
    t.goto(a/2*(n-1),  0)
    x = t.xcor()
else:
    t.goto((n//2)*a, 0)    
    x = t.xcor()
t.dot(10, 'red')
t.pencolor('green')

for i in range(0, n):
    t.goto(a*i, -100)
    t.pendown()
    t.dot(10, 'blue')
    t.goto(x, 0)
    
t.mainloop()
```

**Задание 10.** Напишите программу, которая рисует изображение символа олимпиады в соответствии с образцом.

```python
import turtle as t
r = 50
h = 12
t.hideturtle()
t.speed(0)
t.pensize(10)

dic = {'DeepSkyBlue2': (0, 0), 
       'black': (2*r + h, 0), 
       'red': ((2*r + h)*2, 0), 
       'yellow':((2*r + h)/2, -r -h), 
       'green2': (3*(2*r + h)/2 , -r - h)}

for i in dic:
    t.pencolor(i)
    t.goto(dic[i])
    t.pendown()
    t.circle(r)
    t.penup()
    
t.mainloop()
```

**Задание 11.** Напишите программу, которая рисует изображение мишки в соответствии с образцом.

```python
import turtle as t
t.hideturtle()
t.speed(0)
t.pensize(3)
r = [30, 80, 50, 10]
parts = {1: {'pos': (0, 0), 'size': r[0]},
         2: {'pos': (2.6*r[0]*2, 0), 'size': r[0]},
         3: {'pos': (2.6*r[0], -1.56*r[1]), 'size': r[1]},
         4: {'pos': (2.6*r[0], -1.56*r[1]), 'size': r[2]},
         5: {'pos': (2.6*r[0], -r[2]), 'size': r[3]}}

for i in parts:
    t.goto(parts[i]['pos'])
    t.pendown()
    t.circle(parts[i]['size'])
    t.penup()
t.pendown()
t.right(90)
t.forward(25)
t.penup()
t.goto((1.2*r[0], -0.7*r[0]))
t.pendown()
t.dot(20)
t.penup()
t.goto((2.6*r[0]*2 - 1.2*r[0], -0.7*r[0]))
t.dot(20)
t.mainloop()
```

**Задание 12.** Напишите программу, которая случайным образом рисует снежинки разного цвета и размера в соответствии с образцом.

```python
import turtle as t
import random
import math
t.tracer(0)
t.Screen().colormode(255)
x_screen, y_screen = 800, 800

t.Screen().screensize(x_screen, y_screen) 
t.Screen().setup(x_screen, y_screen)
(x0, y0) = t.screensize()
t.Screen().setworldcoordinates(-x0,-y0, x0, y0)
t.Screen().bgcolor("cyan") 
t.speed(0)

snow = 0 
snow_min_size = 10
snow_max_size = 70
koeff = 1.5
x1, x2 = int(-x0 + koeff * snow_max_size), int(x0 - koeff * snow_max_size)
y1, y2 = int(-y0 + koeff * snow_max_size), int(y0 - koeff * snow_max_size)

Access_x = set(range(x1, x2))  # Доступные точки
Access_y = set(range(y1, y2))

# Рисуем ограничивающую область
# t.penup()
# t.goto(x1, y1)
# t.pendown()
# t.goto(x2, y1)
# t.goto(x2, y2)
# t.goto(x1, y2)
# t.goto(x1, y1)

t.penup()
t.hideturtle()
placed_points = set()

def snowflake(size, pos, snow_color): # рисуем снежинку
    t.speed(0)
    t.pencolor(snow_color)
    t.penup()
    t.goto(pos)
    t.pendown()
    
    if size > 30:
        t.pensize(3)
    elif 15 < size <= 30:
        t.pensize(2)
    else:
        t.pensize(1)
    for _ in range(8):
        t.forward(2/5 * size)
        for _ in range(3):
            t.left(45)
            t.forward(2/5 * size)
            t.backward(2/5 * size)

            t.right(90)
            t.forward(2/5 * size)
            t.backward(2/5 * size)

            t.left(45)
            t.forward(2/5 * size)
        t.backward(8/5 * size)
        t.left(360/8)

def is_valid(pos, size):
    global placed_points

    border_x = set(range(math.ceil(pos[0] - 8/5*size), math.ceil(pos[0] + 8/5*size + 1)))
    border_y = set(range(math.ceil(pos[1] - 8/5*size), math.ceil(pos[1] + 8/5*size + 1)))
    flag = True

    for i in border_x:
        for j in border_y:
            if (i, j) in placed_points:
                flag = False
                break
        if flag == False:
            break
      
    if flag == True:
        for i in border_x:
            for j in border_y:
                placed_points.add((i, j))     
    return flag
    
def point_position(Access_x, Access_y):
    coordinates = {} 
    x0 = random.choice(tuple(Access_x))  # Выбор коор. из доступных точек
    y0 = random.choice(tuple(Access_y)) 
    pos = (x0, y0) # центр снежинки

    coordinates[pos] =  coordinates.get(pos, 0) + 1

    while coordinates[pos] > 1:
        x0 = random.choice(tuple(Access_x))  
        y0 = random.choice(tuple(Access_y)) 
        pos = (x0, y0)
        coordinates[pos] =  coordinates.get(pos, 0) + 1
        if coordinates[pos] == 1:
            break
    return pos    
            
def drawing():
    global snow, snow_max_size
    global Access_x, Access_y
    global col
    pos = point_position(Access_x, Access_y)
    current_pos = pos
    size = random.randrange(snow_min_size, snow_max_size)
    res = is_valid(current_pos, size) 
    color = random.sample([c for c in range(1, 256)], 3)
    
    if res:
        snow += 1
        # print('СНЕЖИНКА', snow)
        snowflake(size, current_pos, color)
    else:
        num = 0 # кол-во попыток генераций снежинки
        while res is False:
            if num <= 100:
                pos = point_position(Access_x, Access_y)
                current_pos = pos
                size = random.randrange(snow_min_size, snow_max_size)
                res = is_valid(current_pos, size)
                num += 1
            elif 100 < num <= 500:
                pos = point_position(Access_x, Access_y)
                current_pos = pos
                size = int(random.uniform(0.2, 0.8) * random.randrange(snow_min_size, snow_max_size - 20))
                res = is_valid(current_pos, size)
                num += 1
            elif num > 500:
                res = True
        snow += 1
        col.append(num)
        # print('СНЕЖИНКА', snow,'num:', num)
        snowflake(size, current_pos, color)
     
n = 800
print(f'Количество снежинок: {n}')   
col = []
for _ in range(n):       
    drawing()

print(max(col), '<', 500) if max(col) < 500 else print('Второй слой!') # если < 500, то снежинки не размещались вторым слоем

t.mainloop()
```
