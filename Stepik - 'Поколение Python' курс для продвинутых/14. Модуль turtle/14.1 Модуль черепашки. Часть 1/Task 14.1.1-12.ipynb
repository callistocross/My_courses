**Задание 1.**
Напишите программу, которая рисует прямоугольник.

```python
import turtle
turtle.pensize(2)

def rectangle(width, height):
  for _ in range(4):
    turtle.forward(width)
    turtle.left(90)
    width, height = height, width

# width, height = int(input()), int(input())
width, height = 200, 100
rectangle(width, height)
turtle.mainloop()
```

**Задание 2.**
Напишите программу, которая рисует правильный треугольник.

```python
import turtle
turtle.pensize(2)

def triangle(side):
  for _ in range(3):
    turtle.forward(side)
    turtle.left(120)
# side = int(input())
side = 200
triangle(side)
turtle.mainloop()
```

**Задание 3.**
Напишите программу, которая рисует изображенную фигуру, состоящую из трех квадратов.

```python
import turtle
turtle.pensize(2)

def square(side):
  for num in [30, 45, 60]:
    turtle.setheading(num)
    for _ in range(4):
      turtle.forward(side)
      turtle.left(90)
  
# side = int(input())
side = 100
square(side)
turtle.mainloop()
```
