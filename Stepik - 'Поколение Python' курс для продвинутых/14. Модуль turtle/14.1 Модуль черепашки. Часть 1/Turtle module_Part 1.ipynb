**Задание 1.**
Напишите программу, которая рисует прямоугольник.

```python
import turtle
turtle.pensize(2)

def rectangle(width, height):
  for _ in range(4):
    turtle.forward(width)
    turtle.left(90)
    width, height = height, width

# width, height = int(input()), int(input())
width, height = 200, 100
rectangle(width, height)
turtle.mainloop()
```

**Задание 3.**
Напишите программу, которая рисует изображенную фигуру, состоящую из трех квадратов.

```python
import turtle
turtle.pensize(2)

def square(side):
  for num in [30, 45, 60]:
    turtle.setheading(num)
    for _ in range(4):
      turtle.forward(side)
      turtle.left(90)
  
# side = int(input())
side = 100
square(side)
turtle.mainloop()
```

**Задание 8.**
Напишите программу, которая рисует снежинку из 10 ромбов.

```python
import turtle
turtle.pensize(2)
n = 10 # кол-во ромбов
def rhombus(side, n):
    turtle.left(360 / n) 
    a, b = 50, 130
    for num in [a, b]*2:
        turtle.forward(side)
        turtle.left(num)          
side = 50
for _ in range(n):
    rhombus(side, n)
turtle.mainloop()
```

**Задание 9.**
Напишите программу, которая рисует лучи звезды, показанной на рисунке.

```python
import turtle
turtle.pensize(2)
side = int(input())
angle = 30
for _ in range(360//angle):
    turtle.forward(side)
    turtle.backward(side)
    turtle.left(angle)
turtle.mainloop()
```

**Задание 11.**
Напишите программу, которая рисует квадраты, чтобы создать узор, показанный на рисунке.

```python
import turtle
turtle.pensize(2)
side = 20

def square(side):
  for _ in range(4):
      turtle.left(90)
      turtle.forward(side)
    
for _ in range(15):
    square(side)
    side += 10
turtle.mainloop()
```
