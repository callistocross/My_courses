**Задание 1.**
Напишите программу, которая рисует изображение домика по образцу.

```python
import turtle as t
side = 100
t.speed(0)
t.pensize(2)
t.hideturtle()
t.fillcolor('DeepSkyBlue')
t.begin_fill()
for _ in range(4):
    t.forward(side)
    t.left(90)
t.end_fill()
t.penup()
t.goto(-side / 4, side)
t.pendown()
t.fillcolor('sienna')
t.begin_fill()
for _ in range(3):
    t.forward(side + side / 2)
    t.left(120)
t.end_fill()
t.mainloop()
```

Напишите программу, которая рисует изображение светофора по образцу.

```python
import turtle as t
side = 300
t.speed(0)
t.pensize(2)
t.hideturtle()
t.fillcolor('Black')
t.begin_fill()

for i in range(4):
    if i % 2 == 0:
        t.forward(side / 2)
        t.left(90)
    else:
        t.forward(side)
        t.left(90)
t.end_fill()

r = side / 8
for color in ['green', 'yellow', 'red']:
    t.penup()
    if color == 'green':
        t.goto(side / 4, 0.5 * ((side / 2) - r) - r)
    elif color == 'yellow':
        t.goto(side / 4, (side / 2) - r)
    elif color == 'red':
        t.goto(side / 4, 1.5 * ((side / 2) - r)  + r)
    t.pendown()
    t.fillcolor(color)
    t.begin_fill()
    t.circle(r)
    t.end_fill()

t.mainloop()
```

```python
import turtle as t
import random
x_screen, y_screen = 500, 300
t.Screen().screensize(x_screen, y_screen)  
t.Screen().setup(x_screen, y_screen)
(x0, y0) = t.screensize()

t.Screen().setworldcoordinates(-x0,-y0, x0, y0)
t.Screen().colormode(255)
t.Screen().bgcolor('DodgerBlue4')

t.Screen().tracer(1, 0)
t.hideturtle()

min_size, max_size = 20, 70
Access_x = set(range(-x0 + max_size, x0 - max_size, 5)) 
Access_y = set(range(-y0 + max_size, y0 - max_size, 5))

def drawing_star():
    t.penup()
    t.goto(random.choice(tuple(Access_x)), random.choice(tuple(Access_y)))
    t.pendown()
  
    coloring = random.randrange(100, 255), random.randrange(100, 255), random.randrange(100, 255)
    t.color(coloring)
    t.fillcolor(coloring)
    size = random.randrange(min_size, max_size)

    t.left(random.randrange(0, 180))
    t.begin_fill()
    for _ in range(5):
        t.forward(size)
        t.right(144)
    t.end_fill()
  
for _ in range(75):
    drawing_star()

t.mainloop()
```

**Задание 8.**
Напишите программу, которая рисует изображение правильных многоугольников по образцу. Многоугольники должны иметь разный цвет. 

Примечание 1. Доработайте программу, чтобы площадь всех многоугольников была одинаковой. Площадь правильного многоугольника вычисляется по формуле 
$S =\cfrac{na^2}{4tg(\frac{180^{\circ}}n)}$, где $n$−количество сторон, $a$−длина стороны.

```python
import turtle as t
from random import randint
from math import sqrt, radians, tan

t.Screen().tracer(1, 0)
t.Screen().colormode(255)
x_screen, y_screen = 300, 300
t.Screen().screensize(x_screen, y_screen)  
t.Screen().setup(x_screen, y_screen)

def draw_figure(n, m):
    global S
    (x0, y0) = t.screensize()
    t.Screen().setworldcoordinates(-x0, -y0, x0, y0)
    s_x =  2 * x0 / n
    s_y =  2 * y0 / m
    for i in range(m):
        for j in range(n):
            num_sides =  randint(3, 8)
            a = sqrt((4 * S * tan(radians(180/num_sides))) / num_sides)
            coord_x = ((-x0 + j * s_x) + (-x0 + (j + 1) * s_x)) / 2
            d = 50 # поправка для высоты
            coord_y = ((y0 - i * s_y) + (y0 - (i + 1) * s_y)) / 2 + d
            t.penup()
            t.goto(coord_x, coord_y)
            t.pendown()
            t.pensize(2)
            t.fillcolor(tuple((randint(0, 255) for _ in range(3))))
            t.begin_fill()
            t.forward(a / 2)
            t.right(360 / num_sides)
            for _ in range(num_sides - 1):
                t.forward(a)
                t.right(360 / num_sides)
            t.forward(a / 2)
            t.end_fill()
            print(f'num_sides = {num_sides}, a = {round(a, 2)}, S = {round(S, 3)}')
            
n, m = 5, 5 # количество фигур
S = 3000
draw_figure(n, m)
       
t.hideturtle()
t.mainloop()
```

**Задание 9.**
Напишите программу, которая рисует изображение шахматной доски по образцу.

```python
import turtle as t
from random import randint
from math import sqrt, radians, tan

t.Screen().tracer(1, 0)
t.Screen().colormode(255)
x_screen = 200
y_screen = x_screen
t.Screen().screensize(x_screen, y_screen)  
t.Screen().setup(x_screen, y_screen)

def draw_square(n):
    global S
    (x0, y0) = t.screensize()
    t.Screen().setworldcoordinates(-x0, -y0, x0, y0)
    d = 30
    step =  2 * (x0 - d) / n
    numb = 0
    for i in range(n):
        for j in range(n):
            coord_x = ((-x0 + j * step) + (-x0 + (j + 1) * step) + d) / 2 
            coord_y = ((y0 - i * step) + (y0 - (i + 1) * step) - d) / 2 - step/2
            t.penup()
            t.goto(coord_x, coord_y)
            t.pendown()
            t.pensize(2)
            t.fillcolor('black') if numb % 2 == 0 else t.fillcolor('white')
            t.begin_fill()
            t.forward(step / 2)
            t.left(90)
            for _ in range(3):
                t.forward(step)
                t.left(90)
            t.forward(step / 2)
            t.end_fill()
            numb += 1
                      
n = 5
draw_square(n)
       
t.hideturtle()
t.mainloop()
```

**Задание 12.**
Напишите программу, которая рисует знак STOP по образцу.

```python
import turtle as t
from math import tan, pi, sin
t.Screen().tracer(1, 0)

t.Screen().colormode(255)
x_screen, y_screen = 400, 500
t.Screen().setup(x_screen, y_screen)
(x0, y0) = t.screensize()
t.Screen().setworldcoordinates(-x0, -y0, x0, y0)

def draw_figure(r, color): 
    t.fillcolor(color)
    t.begin_fill()
    a = 2 * r * tan(pi / 8)
    t.forward(a / 2)
    t.left(360 / 8)
    for _ in range(7):
        t.forward(a)
        t.left(360 / 8)
    t.forward(a / 2)   
    t.end_fill()  

r1 = 200
r2 = r1 - 10
r3 = r2 - 10
draw_figure(r1, 'black')   
h = r1 - r2  
t.penup()   
t.goto(0, h)  
draw_figure(r2, 'white')
h = h + r2 - r3
t.penup()
t.goto(0, h) 
draw_figure(r3, 'red')
t.penup()
w = 46
t.goto(-r3 + 14, r1 / 2 + w / 2 + 10) 
t.pencolor('white')
t.write('STOP', font=('Arial', w, 'normal'))  # печатаем надпись STOP
t.hideturtle()
t.mainloop()    
```

**Задание 13.**
Напишите программу, которая рисует силуэты многоэтажек по образцу. Разделите программу на функции:
- рисования контуров зданий;
- рисования нескольких окон в зданиях;
- рисования случайно разбросанных звезд в виде точек (убедитесь, что звезды появляются на небе, а не на зданиях).

```python
import turtle as t
import random
t.Screen().tracer(1, 0)

t.Screen().colormode(255)
t.Screen().bgcolor('MidnightBlue')
x_screen, y_screen = 1200, 600
t.Screen().setup(x_screen, y_screen)
(x0, y0) = t.screensize()
t.Screen().setworldcoordinates(-x0, -y0, x0, y0)

min_size, max_size = 10, 15
Access_x = set(range(-x0 + max_size, x0 - max_size, 5)) 
Access_y = set(range(-y0 + 400 - max_size, y0 - max_size, 5))

def move_to(x, y):
    t.penup()
    t.goto(x, y)
    t.pendown()
    
def drawing_star(count):
    for _ in range(count):
        move_to(random.choice(tuple(Access_x)), random.choice(tuple(Access_y)))
        coloring = 'AliceBlue'
        t.color(coloring)
        t.fillcolor(coloring)
        size = random.randrange(min_size, max_size)
        t.left(random.randrange(0, 180))
        t.begin_fill()
        for _ in range(5):
            t.forward(size)
            t.right(144)
        t.end_fill()
    t.setheading(0)

def building(a, b, color, command = False):
    t.color('black')  
    numb = random.randrange(100)
    if command:
        if numb % 2 == 0:
            t.fillcolor(color)
        else:
            t.fillcolor('gray50')
        t.begin_fill()
        for _ in range(2):
            t.forward(a)
            t.left(90)
            t.forward(b)
            t.left(90)
        t.end_fill()
    else:
        t.fillcolor(color)
        t.begin_fill()
        for _ in range(2):
            t.forward(a)
            t.left(90)
            t.forward(b)
            t.left(90)
        t.end_fill()
           
count = 1
def draw_building():
    global count
    y = -y_screen
    if count == 1:
        x = -x_screen
    while x < x0:
        a = random.randrange(300, 600, 50)
        b = random.randrange(400, 1100, 20)
        move_to(x, y)
        if t.xcor() + a > x_screen:
            a = x_screen - t.xcor()
        building(a, b, 'LightSteelBlue3')
        x = t.xcor() + a 
        count += 1
        
        if   700 <= b and 500 <= a <= 600:
            window, p, max_count, d, r = 25, 40, 10, b / 10, 4
        elif 700 <= b and 300 <= a < 500:
            window, p, max_count, d, r = 25, 60, 10, b / 10, 3
        elif 700 <= b and 150 <= a < 300:
            window, p, max_count, d, r = 25, 40, 10, b / 10, 2
        elif 700 <= b and a < 150:
            window, p, max_count, d, r = 15, 40, 10, b / 10, 1  
        elif 400 <= b < 700 and 500 <= a <= 600:
            window, p, max_count, d, r = 25, 40, 6, b / 6, 4 
        elif 400 <= b < 700 and 300 <= a < 500:
            window, p, max_count, d, r = 25, 60, 6, b / 6, 3 
        elif 400 <= b < 700 and 150 <= a < 300:
            window, p, max_count, d, r = 25, 40, 6, b / 6, 2 
        elif 400 <= b < 700 and a < 150:
            window, p, max_count, d, r = 15, 40, 6, b / 6, 1 
        
        for i in range(1, max_count):
                color_light = 'bisque'
                if r == 1:
                    move_to((x - a + p + x - p - window)/2, -y_screen + i * d)
                    building(window, window, color_light, True)
                elif r == 2:
                    move_to(x - a + p, -y_screen + i * d)
                    building(window, window, color_light, True)
                    
                    move_to(x - p - window, -y_screen + i * d)
                    building(window, window, color_light, True)
                elif r == 3:
                    move_to(x - a + p, -y_screen + i * d)
                    building(window, window, color_light, True)

                    move_to((x - a + p + x - p - window)/2, -y_screen + i * d)
                    building(window, window, color_light, True)
                    
                    move_to(x - p - window, -y_screen + i * d)
                    building(window, window, color_light, True)
                elif r == 4:
                    move_to(x - a + p, -y_screen + i * d)
                    building(window, window, color_light, True)

                    move_to((x - a + p + x - p - window)/2 - window - p, -y_screen + i * d)
                    building(window, window, color_light, True)
                    
                    move_to((x - a + p + x - p - window)/2 + window + p, -y_screen + i * d)
                    building(window, window, color_light, True)
                    
                    move_to(x - p - window, -y_screen + i * d)
                    building(window, window, color_light, True)
drawing_star(150)                
draw_building()    

t.hideturtle()
t.mainloop()  
```

**Задание 14.**
Напишите программу, которая рисует изображение сердца по образцу.

```python
import turtle as t
from math import pi, sin, cos
t.Screen().tracer(1, 0)

t.Screen().colormode(255)
x_screen, y_screen = 200, 250
t.Screen().setup(x_screen, y_screen)
(x0, y0) = t.screensize()
t.Screen().setworldcoordinates(-x0, -y0, x0, y0)

def heart():
    max_count = int(round(2 * pi, 2) * 100 + 1)
    t.pencolor('red')
    t.fillcolor('red')
    t.begin_fill()
    for i in range(0, max_count):
        i *= 0.01
        x = 128 * sin(i)**3
        y = 8 * (13 * cos(i) - 5 * cos(2 * i) - 2 * cos(3 * i) - cos(4 * i) - 5)
        t.goto(x, y)
    t.end_fill()
    
heart()

t.hideturtle()
t.mainloop()  
```

**Задание 15.**
Напишите программу, которая по нажатию на левую кнопку мыши рисует звезду в месте клика. Фон изображения должен быть черным, при этом звезды могут иметь разные размеры, цвета и иметь разное количество сторон.

```python
import turtle as t
from math import cos, sin, pi
import random
t.Screen().tracer(1, 0)

t.Screen().colormode(255)
t.Screen().bgcolor('Black')
x_screen, y_screen = 300, 300
t.Screen().setup(x_screen, y_screen)
(x0, y0) = t.screensize()
t.Screen().setworldcoordinates(-x0, -y0, x0, y0)

def left_mouse_click(x, y):
    star(x, y)

def star(x, y):
  t.penup()
  t.setpos(x, y)
  t.pendown()
  n = random.randint(3, 10)
  size = random.randint(10, 30)
  color = (random.randrange(256), random.randrange(256), random.randrange(256))
  t.setheading(random.randint(0, 361))
  t.pencolor(color)
  t.fillcolor(color)
  t.begin_fill()
  if n > 3:
    for _ in range(n):
      t.forward(size)
      t.right(180 - 180 / n)
      t.forward(size)
      t.left(180 - 540 / n)
  else:  
    for _ in range(n):
      t.forward(size)
      t.right(150)
      t.forward(size)
      t.left(30)
  t.end_fill()

t.Screen().onclick(left_mouse_click)
t.Screen().listen()
t.hideturtle()
t.mainloop()  
```
